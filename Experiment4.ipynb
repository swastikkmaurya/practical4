```python
'''Problem Statement 2 : three students has choose elective subjects
Find the commom subjects of all three students by set operations.

```

```python
stud1 = {"python","aiml","entc","ra","civil"}
stud2 = {"python","aiml","sports","marketing","cullinary"}
stud3 = {"python","aiml","java","c++","human values","cyber security"}

common_subjects = stud1 & stud2 & stud3
print("Common Subjects : ",common_subjects)
```
